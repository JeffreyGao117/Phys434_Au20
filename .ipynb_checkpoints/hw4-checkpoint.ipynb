## Problem 0


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

fig1, ax1 = plt.subplots(1,1, figsize=(15, 5))

t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)  # offset sine
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
ax1.scatter(t, bkgd, 2);
```

## Problem 1 - Array Slicing

```python
fig2, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,10))

#----- 1A

t1 = t[:1000]
bkgd1 = bkgd[:1000]
ax1.plot(t1, bkgd1, 'ro', markersize=2);
ax1.set_title('First 1000 Points')

#----- 1B

start = 500
stop = 2000
t2 = t[start:stop]
bkgd2 = bkgd[start:stop]
ax2.plot(t2, bkgd2, 'ro', markersize=2);
ax2.set_title('Points 500 to 2000')


#----- 1C

step = 100
t3 = t[0:-1:step]
bkgd3 = bkgd[0:-1:100]
ax3.scatter(t3, bkgd3, 'ro', markersize=2);
ax3.set_title('Every 100th Point')

plt.show()
```

## Problem 2 - Density

```python
#----- 2A

fig3 = plt.figure(figsize=(15,10))
fig3 = plt.hist2d(t, bkgd, bins=100)
plt.colorbar(fig3[3])

plt.show()
```

__2B)__ The 2D histogram is plotting th density of points at along each x-position. The light green color indicates a higher density of points while the bluer colors refer to less point density or no points at all. The density is represented by the quantity of points within a bin width.

## Problem 3 - Folding

```python
#----- 3A

tf = np.linspace(0,np.pi*100,np.int(1e5))  % (2*np.pi)  # folded
Nf = 10+np.sin(tf)  # offset sine
bkgdf = stats.norm.rvs(size=np.int(1e5))*np.sqrt(Nf)+Nf
fig4 = plt.figure(figsize=(15,5))
fig4 = plt.scatter(tf, bkgdf, 2);
plt.title('Points with folded time variable')

#----- 3B

fig5 = plt.figure(figsize=(15,10))
fig5 = plt.hist2d(tf, bkgdf, bins=100)
plt.colorbar(fig5[3])

#----- 3C

avg = []

start = 0
for n in range(1,101):
    stop = n * 1000
    avg.append(np.average(bkgdf[start:stop]))
    start = stop

fig5 = plt.plot(fig5[1][0:100], avg, 'ro')

plt.show()
```
